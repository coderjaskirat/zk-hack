```python
from onnxruntime import InferenceSession
import numpy as np
import onnx
import os
```

# Usage example classifying model

```python
filename = "/Users/lnm/montreal/zk-hack/ezkl-example/models/mobilenetv2-1.0.onnx"
print(os.getcwd())
```

```python
# explore input naming and formats
model = onnx.load(filename)
onnx.checker.check_model(model)
inputs = {}
outputs = {}
for inp in model.graph.input:
    shape = str(inp.type.tensor_type.shape.dim)
    inputs[inp.name] = [int(s) for s in shape.split() if s.isdigit()]
for outp in model.graph.output:
    shape = str(outp.type.tensor_type.shape.dim)
    outputs[outp.name] = [int(s) for s in shape.split() if s.isdigit()]
#print(inputs)
print(outputs)
```

```python
def execute_onnx_model_from_file(filename: str, image_array: np.array) -> None:
    sess = InferenceSession(filename)

    res = sess.run(None, {"data": image_array})[0]
    print(res)
```

```python
import json
dummy_input = np.random.rand(1, 3, 224, 224).astype(np.float32)

data = {'input_data': [dummy_input.reshape([-1]).tolist()]}
with open('input.json', 'w') as f:
    json.dump(data, f)

execute_onnx_model_from_file(filename, dummy_input)
```
